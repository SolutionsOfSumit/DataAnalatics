to do anything with the file which we have we first have to open it

```python
os.listdir("./data")
```

```python
file = open('./data/data.txt', mode= 'r')
```

```python
file_contents = file.read()
print(file_contents)
```

```python
file.close()
```

the `.close()` wil close the file which should be done everytime when you are done using the file you have opened or it will eat up the ram space in you computer. after closing the file the file cannot be read but the contents still exist in file_contents

Now to ease this proces we use `woth` statement to open and close the file sametime after we are done with the file

```python
with open('./data/data.txt') as file2:
    file2_contents = file2.read()
    print(file2_contents)
```

We can even read file line by line for specific usecases

```python
with open("./data/data.txt") as file3:
    file3_contents = file3.readlines() #this will give lines as a particular string
    print(file3_contents)
```

```python
file3_contents[0]
```

You can see a newline character after our string which can be removed by using the following code
