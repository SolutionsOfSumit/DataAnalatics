Now we have a list of keys which we can use

```python
def parse_values(data_line):
    values = []
    for item in data_line.strip().split(","):
        values.append(float(item))
    return values
```

just because we had a "," in the end of the last line it has generated a empty string in the end of the list. which can create problems if we have missing values in our dataset.

In our data set line 4 and 6 will be having missing values

If that is the case we have to modify our function to handel such case.

```python
# this is called handling  missing values

def parse_values(data_line):
    values = []
    for item in data_line.strip().split(","):
        if item == "":
            values.append(0.0)
        else:
            values.append(float(item))
    return values
```

Here we can clearly see that the 0.0 value got inserted in the place of missing value

```python
def create_item_dict(values, headers):
    result = {}
    for values, header in zip(values, headers):
        result[header] = values
    return result
```

```python
def csv_to_dict(path):
    result = []
    with open(path, "r") as file:
        lines = file.readlines()
        headers = parse_headers(lines[0])
        for data_line in lines[1:]:
            values = parse_values(data_line)
            item_dict = create_item_dict(values, headers)
            result.append(item_dict)
    return result
```

```python
new_data = csv_to_dict("./data.txt")
```

```python
import math
def loan_emi(amount, duration, rate, down_payment=0):
    """
    Calculates the Equal Monthly Installment (EMI) for a loan.

    Arguments:
        amount         - Total cost of the asset.
        duration       - Loan duration in months.
        rate           - Annual interest rate (e.g. 10 for 10%).
        down_payment   - Initial payment (optional).

    Returns:
        Monthly EMI.
    """
    loan_amount = amount - down_payment
    monthly_rate = rate / (12 * 100)

    try:
        emi = (
            loan_amount
            * monthly_rate
            * (1 + monthly_rate) ** duration
            / ((1 + monthly_rate) ** duration - 1)
        )
        return emi
    except ZeroDivisionError:
        return loan_amount / duration
```
