# While while loop is used based on a statement but for loops are used if the sequence is satisifed or not 

````python
days = [ "Monday" , "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"] 

for day in days:
    print(day)
````

````python
person1 = dict(name="John doe", sex="male", age=22, city="Hyd")
for key in person1:
    print("Key:", key, "," , "Value:", person1[key])
````

# Range is little tricker 
```python 
for a in range(a, b, c) #here a is the starting of the range, b is the ending of the range and c is teh difference between the iterations also known as step 
```
Bellow are some examples for better understanding

````python
for i in range(3,10): # This loop will print form 3 all the way up to 9 because end ending range will not get printed
    print(i)
````

````python
# Bellow loop is an example of how to use step 
for i in range (3,12,4):
    print(i)
````

Here we can see how the stesps work in every iteration loop skiped 4 numbers, this also work with negative statement as the making the trickey question reversing a string work. Here is one example bellow

````python
for i in range (10, 1, -1):
    print(i)
````

Now the infamous quesiton of reversing a string

````python
string = "monster"
result = ""
for i in range(len(string) -1, -1, -1):
    result += string[i]

print(result)
````
