# Working with CSV files which stands for Comma Separated Values format

we will also be using urllib which is used to import urls in our document

```python
import urllib.request as url

url.urlretrieve(
    'http://hub.jovian.ml,wp-content/uploads/2020/08/climate.csv','cliate.txt'
)
```

```python
climate_data = np.genfromtxt('climate_simple.txt', delimiter=',', skip_header=1)
```

```python
climate_data
```

```python
climate_data.shape
```

```python
weights = np.array([0.3, 0.2, 0.5])
yields = climate_data @ weights
yields
```

```python
yields.shape
```

we want to add this new data to the file climate_sample.txt as a new element in each row and a new column in the exesting matrix

```python
climate_result = np.concatenate((climate_data, yields.reshape(10000, 1)), axis=1)
```

```python
climate_result
```

```python
np.savetxt('climate_result.txt',
            climate_result,
            fmt='%.2f',
            comments='',
            header='temperature,rainfall,humidity,yeild_apples'
)
```
