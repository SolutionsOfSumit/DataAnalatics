# Analyzing data form data frames 
Till now we are done with basics of pandas now we can move on to analyze the data and finding out revelent and figuering our revelent information.

````python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
````

````python
print('The number of reportd cases is {} and the number of reported deaths{}'.format(total_cases, total_deaths))
````

*What is the overall number of tests conducted? A total of 935310 tests were conducted before daily test number being reported*

````python
initial_tests = 935310
total_tests = initial_tests + covid_df2.new_tests.sum()
````

````python
total_tests
````

````python
positive_rate = total_cases / total_tests
````

````python
print('{:.2f}% of tests in Italy led to positive diagonsis.'.format(positive_rate))
````

The data contains 8078 rows but only first and last 5 are visiable by default. to view all the rows, we can modify some display options 

````python
from IPython.display import display

with pd.option_context('display.max_rows', 100):
    display(covid_df[covid_df.new_cases > 1000])
````

The above code is not showing the hundread lines because.```display.max_rows``` This only sets the maximum number of rows that Pandas is allowed to display. To see atleast 100 we have to set ```display.min_rows``` as 100

````python
from IPython.display import display

with pd.option_context(
    'display.max_rows', 100,
    'display.min_rows', 100
):
    display(covid_df[covid_df.new_cases > 1000])
````

````python
positive_rate # This shows what percentage of cases turned out to be positive in compression to all of them
````

````python
high_ratio_df = covid_df[covid_df.new_cases / covid_df.new_tests > positive_rate]
````

Here we are dividing a series by series. This will give you element wise divisions.

````python
high_ratio_df
````

this will give you a new series where ratio of new cases were higher then the new tests.

**Further you can us this series to add a new column to the data frame.**

````python
covid_df2["positive_rate"] = covid_df2.new_cases / covid_df2.new_tests
````
